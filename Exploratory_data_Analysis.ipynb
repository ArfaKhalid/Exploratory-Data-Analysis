# Exploratory Data Analysis:
Exploratory Data Analysis (EDA) is a method used to `analyze` and `summarize` datasets. In Python, you can use the pandas library to perform EDA on a dataset.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# summarize (summary statistics)
df.describe()
```

```python
# find null values True = null values present
df.isnull()
```

```python
df.isnull().sum()
```

```python
# Datawrangling -  deals with missing values
df.isnull().sum() * 100 /len(df) # length function gives the length of the data
```

```python
#EDA
#finding unique values
df.nunique()
```

```python
df.columns
```

```python
# visualize the nan value in heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
#Data Manipulation In EDA
df.drop(columns="deck", inplace=True)
```

```python
df.shape
```

```python
#Visulaize and explore
sns.countplot(data=df, x="survived")
```

```python
sns.barplot(data=df, x="survived", y="age", hue="who")
```

```python
sns.countplot(x="pclass", data=df)
```

```python
# data distriibution
sns.histplot(data=df, x="age")
```

```python
#Visulaize data distribution by making groups using bins
#categorical data
sns.histplot(data=df, x="age", bins=30)
```

```python
# Gradient (use alpha)
#numerical data
df['age'].hist(bins=30, color='darkred', alpha=0.75)
```

```python
#for categorical and numerical data
sns.boxplot(data=df, y="age", x="survived", hue="who")
```

# Outlier - using box plot
when data are divided into bins we can read it using its quartiles to find outlier


```python
# scatter plot to detect outliers
# it shows that data are distributed greatly which means that data is not normally distributes
sns.scatterplot(x='age', y='fare', data=df, hue="pclass")
plt.ylim(0,300) # it changes the scale in the graph
```

```python
sns.pairplot(df)
```

# Pairplot
it gives information about each variables with each other

```python
sns.pairplot(data=df, hue="survived")
```

```python
#stacked barchat/plot
df.groupby(['pclass', 'who'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
#swarm plot
sns.swarmplot(data=df, x="age", y="fare")
```

```python
sns.swarmplot(data=df, x="survived", y="age", hue="who")
```

```python
#point plot
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df)
```

```python
#relation of two numeric number variables
filtered_df = df[df.fare < 300]
sns.lmplot(data=filtered_df, x='age', y='fare', hue="pclass")
```

# Groups in Data

```python
# how to add a column and make a age group
sns.histplot(data=df, x='age')
```

```python
#grouping
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'], [0,10,18,30,50,80],
                          labels=['Kids', 'Child', 'Youth', 'Middle aged', 'Old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_groups")["survived"].mean().plot(kind="bar")
```

```python
sns.pairplot(df, hue='survived', vars=['age','fare','pclass'])
```
